```python
import pandas as pd
import numpy as np

import warnings
warnings.filterwarnings('ignore')
```

```python
customer_df = pd.read_csv('data/generated_customer.csv')
```

```python
customer_df
```

```python
customer_demographic_dim = pd.DataFrame(columns=['InterestedTopic', 'AccountType'])
customer_demographic_dim
```

```python
for i in customer_df['account_type'].unique():
    for j in customer_df['interested_topic'].unique():
        customer_demographic_dim = customer_demographic_dim.append({'InterestedTopic': j, 'AccountType': i}, ignore_index=True)
```

```python
customer_demographic_dim
```

```python
customer_demographic_dim.to_csv('dimensions/generated_customer_demographic.csv', index=False)
```
