# Convert Sale date to DateTime

```python
import pandas as pd
import numpy as np
```

```python
df= pd.read_csv('./data/generated_amazon_sale_report.csv')
```

```python
df.head()
```

```python
df['date'] = pd.to_datetime(df['date'])

df['date'] = df['date'].dt.strftime('%d-%m-%Y')
df.head()
```

```python
df.to_csv('./data/generated_amazon_sale_report.csv', index=False)
```
