```python
import pandas as pd
import numpy as np
```

## Promotion IDs

```python
sale_df = pd.read_csv('data/datetime_amazon_sale_report.csv')
sale_df
```

```python
sale_df.dropna(inplace=True)
```

```python
promotions = []

for promo in sale_df['promotion_ids'].unique()[0:20]:
    for each in promo.split(','):
        if each not in promotions:
            promotions.append(each)
```

```python
promotions = []

for promo in sale_df['promotion_ids'].unique():
    for each in promo.split(','):
        if each not in promotions:
            promotions.append(each)
```

```python
len(promotions)
```

```python
for promo in promotions:
    print(promo)
```

```python
promotion_dim = pd.DataFrame(data=promotions,columns=['PromotionID'])
promotion_dim
```

```python
promotion_dim.to_csv('dimensions/promotion_dim_file.csv',index=False)
```
