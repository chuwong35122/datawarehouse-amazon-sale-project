```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('./data/datetime_amazon_sale_report.csv')
```

```python
delivery_indicator_dim = pd.DataFrame(columns=['SaleChannel', 'ServiceType', 'ShippingCompany'])
```

```python
customer_dim = pd.read_csv('./data/generated_customer.csv')
```

```python
for channel in df['sales_channel'].unique():
    for acc_type in customer_dim['account_type'].unique():
        for fulfilled in df['fulfilled_by'].unique():
            delivery_indicator_dim = delivery_indicator_dim.append({'SaleChannel': channel, 'ServiceType': acc_type, 'ShippingCompany': fulfilled}, ignore_index=True)
```

```python
delivery_indicator_dim.to_csv('./dimensions/delivery_indicator_dim_file.csv', index=False)
```
